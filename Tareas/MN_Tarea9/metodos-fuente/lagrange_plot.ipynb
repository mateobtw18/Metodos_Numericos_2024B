```python
xs = [1, 2, 3]
ys = [1, 2, 2]
```

```python
import matplotlib.pyplot as plt

plt.scatter(xs, ys)
plt.xlabel("xs")
plt.ylabel("ys")
plt.title("Plot of xs vs ys")
plt.show()
```

```python
def f(x: float) -> float:
    return -(x**2) / 2 + 5 * x / 2 - 1
```

```python
import numpy as np

x_values = np.linspace(min(xs), 4, 100)
y_values = [f(x) for x in x_values]

plt.plot(x_values, y_values, label="f(x)")
plt.scatter(xs, ys, color="red")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.legend()
plt.show()
```

```python
def f2(x: float) -> float:
    return x**2 - 2 * x - 4


xs2 = [-1, 3, 0, 2]
y2 = [f2(x) for x in xs2]
print(y2)
```

```python
# Generate x values for plotting f2
x_values_f2 = np.linspace(-1.5, 3.5, 100)
y_values_f2 = [f2(x) for x in x_values_f2]

# Plot f2 and y points
plt.plot(x_values_f2, y_values_f2, label="f2(x)")
plt.scatter(xs2, y2, color="red", label="y points")
plt.scatter(2, -3.75, color="green")
plt.xlabel("x")
plt.grid()

plt.ylabel("f2(x)")
plt.title("Plot of f2(x) and y points")
plt.legend()
plt.show()
```
