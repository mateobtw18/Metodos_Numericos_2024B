```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, x0, y0, h, n):
    """
    Método de Euler para resolver la EDO dy/dx = f(x, y)
    
    :param f: Función f(x, y) que define la EDO
    :param x0: Valor inicial de x
    :param y0: Valor inicial de y
    :param h: Tamaño del paso
    :param n: Número de iteraciones
    :return: Arreglos con los valores de x e y
    """
    x_vals = np.zeros(n+1)
    y_vals = np.zeros(n+1)
    
    x_vals[0] = x0
    y_vals[0] = y0

    for i in range(n):
        y_vals[i+1] = y_vals[i] + h * f(x_vals[i], y_vals[i])  # Fórmula de Euler
        x_vals[i+1] = x_vals[i] + h  # Avanzar en x
    
    return x_vals, y_vals

# Definimos la ecuación diferencial dy/dx = f(x, y)
def f(x, y):
    return x + y  # Ejemplo: dy/dx = x + y

# Parámetros iniciales
x0, y0 = 0, 1  # Condición inicial y(0) = 1
h = 0.1  # Tamaño de paso
n = 20   # Número de iteraciones

# Ejecutamos el método de Euler
x_vals, y_vals = euler_method(f, x0, y0, h, n)

# Graficamos los resultados
plt.plot(x_vals, y_vals, label="Aproximación de Euler", marker="o")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Método de Euler")
plt.legend()
plt.grid()
plt.show()
```
