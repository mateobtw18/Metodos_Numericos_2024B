# Pregunta 2: Raíces del Polinomio

```python
from scipy.optimize import newton
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x ** 5 - 6 * x ** 4 + 2 * x ** 3 + 20 * x ** 2 - 27 * x + 10

def fder(x):
    return 5 * x ** 4 - 24 * x ** 3 + 6 * x ** 2 + 40 * x - 27
```

Las raices son:

```python
p0 = 4
n1 = newton(f, p0, fprime=fder)

print(f"La raiz aproximada es: {n1} | Método de Newton")
```

```python
p0 = 2
n1 = newton(f, p0, fprime=fder)

print(f"La raiz aproximada es: {n1} | Método de Newton")
```

```python
p0 = 10
n1 = newton(f, p0, fprime=fder)

print(f"La raiz aproximada es: {n1} | Método de Newton")
```

## Gráfica

```python
xs = np.linspace(-3, 5, 100)
ys = f(xs)

xs0 = [-2, 1, 5]
ys0 = [f(x) for x in xs0]

plt.plot(xs, ys)
plt.scatter(xs0, ys0, label="Raices")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', lw=0.8)
plt.title('Raices de la función')
plt.legend()
plt.grid()
plt.show()
```
