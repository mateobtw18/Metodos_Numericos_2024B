```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y
```

```python
def df(x, *args):  # Agregamos *args para recibir los argumentos adicionales
    return 3*x**2 - 6*x + 1
```

```python
from scipy.optimize import newton

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.newton.html
x0 = 100
points = []
x_r = newton(
    f, x0, fprime=df, args=(points, True), maxiter=100
)  # Este no es método de Newton, es secante!
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, y, label='$x^3 - 3x^2 + x - 1$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.scatter(x_r, f(x_r), color='red', zorder=5, label='Root')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $x^3 - 3x^2 + x - 1$')
plt.grid(True)
plt.legend()
plt.show()
```
